```python
import os
from pydub import AudioSegment
```

```python
path = "C:/Users/ktauf/Documents/Github/Final-Project-du/mp3"
out_path = "C:/Users/ktauf/Documents/Github/Final-Project-du/wav"

os.chdir(path)

audio_files = os.listdir()

for file in audio_files:
    name, ext = os.path.splitext(file)
    sound = AudioSegment.from_mp3(file)
    #rename them using the old name + ".wav"
    extract = sound[0:30000]
    export = extract.export(f"{out_path}/{name}.wav", format="wav")
    print(export)
```
