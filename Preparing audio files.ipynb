```python
import numpy as np
import pandas as pd
import librosa
import os
import math
import json
```

```python
# define variables
sample_rate = 22050
n_mfcc = 13
n_fft = 2048
hop_length = 512
num_segments = 10
duration = 30
samples_per_track = sample_rate * duration 
num_samples_per_segment = int(samples_per_track / num_segments)
expected_vectors_per_segment = math.ceil(num_samples_per_segment / hop_length)
```

```python
# create dictionary for data
data = {'species' : [],
        'recording_id' : [],
        'mfcc_0' : [],
        'mfcc_1' : [],
        'mfcc_2' : [],
        'mfcc_3' : [],
        'mfcc_4' : [],
        'mfcc_5' : [],
        'mfcc_6' : [],
        'mfcc_7' : [],
        'mfcc_8' : [],
        'mfcc_9' : []}
```

```python
# change working directory
path = "C:/Users/ktauf/Documents/Github/Final-Project-du/wav"
os.chdir(path)
audio_files = os.listdir()
```

```python
mini_dict ={'species' : [],
        'recording_id' : [],
        'mfcc_0' : [],
        'mfcc_1' : [],
        'mfcc_2' : [],
        'mfcc_3' : [],
        'mfcc_4' : [],
        'mfcc_5' : [],
        'mfcc_6' : [],
        'mfcc_7' : [],
        'mfcc_8' : [],
        'mfcc_9' : []} 


file = audio_files[0]
name, ext = os.path.splitext(file)
filename_components = name.split('-')
species = f"{filename_components[0]}_{filename_components[1]}"
recording_parts = filename_components[2].split('.')
recording_id = recording_parts[0]
mini_dict['species'].append(species)
mini_dict['recording_id'].append(recording_id)
signal, sr = librosa.load(file, sr = sample_rate)

for s in range(num_segments):
    start_sample = num_samples_per_segment * s
    finish_sample = start_sample + num_samples_per_segment
    mfcc = librosa.feature.mfcc(signal[start_sample:finish_sample],
                                sr = sample_rate, n_fft = n_fft, 
                                n_mfcc = n_mfcc, hop_length = hop_length)
    mfcc = mfcc.T
    mini_dict[f'mfcc_{s}'].append(mfcc)
    print(len(mfcc))


mini_dict
```

```python
# loop through files to fill in dictionary
for index, file in enumerate(audio_files):
    name, ext = os.path.splitext(file)
    filename_components = name.split('-')
    species = f"{filename_components[0]}_{filename_components[1]}"
    recording_parts = filename_components[2].split('.')
    recording_id = recording_parts[0]
    data['species'].append(species)
    data['recording_id'].append(recording_id)
        
    print(f"Processing file# {index + 1} of {len(audio_files)} for {species}, {recording_id}")
    
    # load file into librosa
    signal, sr = librosa.load(file, sr = sample_rate)
    
    #process segments
    for s in range(num_segments):
        start_sample = num_samples_per_segment * s
        finish_sample = start_sample + num_samples_per_segment
        mfcc = librosa.feature.mfcc(signal[start_sample:finish_sample],
                                    sr = sample_rate, n_fft = n_fft, 
                                    n_mfcc = n_mfcc, hop_length = hop_length)
        mfcc = mfcc.T
        if len(mfcc) == expected_vectors_per_segment: 
            data[f'mfcc_{s}'].append(mfcc.tolist())
            print(len(mfcc))
```

```python
audio_df = pd.DataFrame(data, columns = ['species', 'recording_id', 'mfcc_0', 'mfcc_1', 'mfcc_2', 'mfcc_3', 'mfcc_4', 'mfcc_5', 'mfcc_6', 'mfcc_7', 'mfcc_8'])

audio_df.to_csv('audio_analysis2.csv')
```
