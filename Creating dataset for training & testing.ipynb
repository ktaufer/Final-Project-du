```python
import os
import librosa
```

```python
# dictionary for storing data
data = {
    'labels' : [],
    'mfcc' : []
}
```

```python
# define constants
n_mfcc = 13
n_fft = 2048
hop_length = 512
sample_rate = 22050
```

```python
# change working directory to one containing music files
path = "C:/Users/ktauf/Documents/Github/Final-Project-du/wav"
os.chdir(path)
audio_files = os.listdir()
```

```python
song, sr = librosa.load('Acrocephalus-arundinaceus-178791.wav', sr = sample_rate)

mfcc = librosa.feature.mfcc(song, sr = sample_rate, n_fft = n_fft, n_mfcc = n_mfcc, hop_length = hop_length)
mfcc = mfcc.T

len(mfcc)
```

```python
for file in audio_files:
    
    #get label for file
    name, ext = os.path.splitext(file)
    filename_components = name.split('-')
    species = f"{filename_components[0]}_{filename_components[1]}"
    data['labels'].append(species)
    print(species)
    #load file into librosa, extract mfcc profile
    signal, sr = librosa.load(file, sr = sample_rate)
    mfcc = librosa.feature.mfcc(signal, 
                                sr = sample_rate,
                                n_fft = n_fft,
                                n_mfcc = n_mfcc,
                                hop_length = hop_length)
    mfcc = mfcc.T
    data['mfcc'].append(mfcc.tolist())    
    print(len(mfcc))
```

```python
import json
json_path = 'c:/users/ktauf/documents/github/final-project-du/data2.json'
with open(json_path, 'w') as f:
    json.dump(data, f)
```
